```python
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
```

```python
with open("data/wiki_us.txt", "r") as f:
    text = f.read()
print(text)
```

```python
doc = nlp(text)
```

### Entitiy Recognition

The ent attribute helps in identifying the entity types in a sentence/document. The identification then helps in the usage of the recognised entites for specific purposes.

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
displacy.render(doc, style = "ent")
```
